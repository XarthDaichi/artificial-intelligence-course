$$
f(x,y) = 2x^2y^2 - 6xy^2 + 3y^2 - 1 \\
g(x,y) = (x-2)^2 + 2(y-1)^2 + x \\
$$

```python
import random
import math
random.seed(666)
```

```python
def f(x,y):
    return 2*x**2*y**2 - 6*x*y**2 + 3*y**2 - 1

def g(x,y):
    return (x - 2)**2 + 2*(y-1)**2 + x
```

```python
def gradient_f(x,y):
    f_x = (4*x - 6)*y**2
    f_y = (4*x**2 - 12*x + 6)*y
    return f_x, f_y

def gradient_g(x,y):
    g_x = 2*x -3
    g_y = 4*(y - 1)
    return g_x, g_y
```

```python
def gradient_descent(gradient, critical_point, f, max_iter=100, epsilon=1e-5, initial=None, lr=0.01):
    if initial is None:
        initial = (random.random(), random.random())
    actual_min = f(*critical_point) # *tuple => spread in javascript
    xn, yn = initial
    for iter in range(max_iter):
        grad_x, grad_y = gradient(xn, yn)
        xn, yn = (xn - lr*grad_x, yn - lr*grad_y)
        estimated_min = f(xn,yn)
        print(f'{iter=:2} aprox_{iter}=({xn},{yn}) {actual_min}==?{estimated_min}')
        if abs(actual_min - estimated_min) < epsilon:
            print(f'*** convergence achieved *** at {iter=:}')
            break
```

```python
gradient_descent(gradient_g, (1.5, 1), g, max_iter=150, lr=0.05)
```

```python
gradient_descent(gradient_f, (1.5,0), f, max_iter=150, lr=0.05)
```

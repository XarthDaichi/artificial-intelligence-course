````python
#### RESPUESTA ALBERTO
from sympy import symbols, diff, solve
import matplotlib.pyplot as plt

x = symbols('x')

def f(x):
    return 2/3 * x ** 3 - 7/4 * x ** 2 + 3/2 * x + 1

def test_fun_instrumentado(start, end, inc):
    graph = []
    graph.append([])
    graph.append([])
    nprange = np.arange(start, end, inc)
    for i in nprange:
        a = i
        graph[0] += [a]
        graph[1] += [f(a)]
    return graph

derivada = diff(f(x), x)

print(f'{derivada}')

puntos_criticos = solve(derivada, x)

print(f'{puntos_criticos}')

segunda_derivada = diff(derivada, x)
extremos = [(p, segunda_derivada.subs(x,p)) for p in puntos_criticos]

minimos = [p[0] for p in extremos if p[1] > 0]
maximos = [p[0] for p in extremos if p[1] < 0]

print("M칤nimos: (", minimos[0], f(minimos[0]), ")")
print("M치ximos: (", maximos[0], f(maximos[0]), ")")

g = test_fun_instrumentado(0, 1.5, 0.01)
plt.plot(g[0], g[1], color="blue", label="f(x)") 
plt.plot(minimos[0], f(minimos[0]), 'ro', label="m칤nimo")
plt.plot(maximos[0], f(maximos[0]), 'go', label='maximo')
plt.legend()
plt.xlabel('x')
plt.ylabel("y")
plt.title("Funci칩n f(x)")
plt.show()
````

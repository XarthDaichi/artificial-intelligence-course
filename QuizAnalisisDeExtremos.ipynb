```python
import random
random.seed(666)
import math
```

```python
def f(x,y):
    return (x - 0.5)**2 + (y - 0.25)**2 + 3
```

```python
def gradient_f(x,y):
    f_x = 2*x - 1
    f_y = 2*y - 0.5
    return f_x, f_y
```

```python

def gradient_descent(gradient, 
                     critical_point, 
                     f, 
                     max_iter=100, 
                     epsilon=1e-5, 
                     initial=None, 
                     lr=0.01):
    if initial is None:
        initial = random.random(), random.random()
    actual_min = f(*critical_point)
    xn, yn = initial
    for iter in range(max_iter):
        grad_x, grad_y = gradient(xn, yn)
        xn, yn = (xn - lr * grad_x, yn - lr*grad_y)
        estimated_min = f(xn, yn)
        print(f"{iter=:2} aprox_{iter} = ({xn=:}, {yn=:}) {actual_min}==?{estimated_min} ")
        if abs(actual_min - estimated_min) < epsilon:
            print(f"*** Convergence achieved *** at {iter=:}")
            break
```

```python
gradient_descent(gradient_f, (0.5, 0.25), f, max_iter=150, lr=0.05)
```
